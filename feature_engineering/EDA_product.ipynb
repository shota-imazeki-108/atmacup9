# 子供

```python
df_for_child = df[(df['サブカテゴリ名']=='子供用')]
df_character_bread = df[(df['部門名']=='ホールセールパン') & (df['サブカテゴリ名']=='キャラクター')]
df_child_choco = df[(df['category_name']=='チョコ・ビスクラ__子供') ]
```

```python
target_words = ['ディズニー', 'アナと雪の女王', 'キティ', 'スヌーピー', 'しまじろう', 'ポケモン', 'ポケットモンスター', 'アンパンマン', 'トーマス', 'プリキュア', 'ドラゴンボール', 'ゲキレンジャー', 'デカレンジャー', 'ゴセイジャー', 'スティッチ', 'チョッパー', 'しんちゃん', 'リラックマ', 'ミッキー', 'ミニー', 'パズル＆ドラゴンズ', 'キョウリュウジャー', 'トッキュウジャー', '妖怪ウォッチ', 'トイ・ストーリー', 'ドラえもん', 'ミニオンズ', 'キュウレンジャー', 'チョコビ', 'プーさん', 'ぷーさん', 'ケロケロ', 'サンリオ', 'しょくぱんまん', 'マリオ', 'ぐでたま', 'マイメロディ', 'キキ＆ララ', 'コナン', 'ピカチュウ']
dfs = []
for word in target_words:
    tmp = df[df['商品名'].str.contains(word)]
    print(word, ':', len(tmp))
    dfs.append(tmp)
character_df = pd.concat(dfs)
```

```python
child_item_df = pd.concat([df_for_child, df_character_bread, df_child_choco, character_df])
child_item_df = child_item_df[['JAN', '商品名', '部門CD', '部門名', 'category_name', 'category_id']]
child_item_df.drop_duplicates(inplace=True)
child_item_df.to_csv('../output/child_item.csv', index=None)
```

```python
import pandas as pd
import numpy as np
```

```python
target_category = [38, 110, 113, 114, 134, 171, 172, 173, 376, 435, 467, 537, 539, 629, 768]
target_category_str = [str(col) for col in target_category]
```

```python
lgb1 = pd.read_csv('../pred/lgb_1.csv')
lgb2 = pd.read_csv('../pred/lgb_2.csv')
lgb3 = pd.read_csv('../pred/lgb_3.csv')
lgb4 = pd.read_csv('../pred/lgb_4.csv')
lgb5 = pd.read_csv('../pred/lgb_5.csv')

lgb = lgb1[target_category_str] * 0.2 +  lgb2[target_category_str] * 0.2 +  lgb3[target_category_str] * 0.2 +  lgb4[target_category_str] * 0.2 +  lgb5[target_category_str] * 0.2
del lgb1, lgb2, lgb3, lgb4, lgb5
```

```python
cat1 = pd.read_csv('../pred/cat_1.csv')
cat2 = pd.read_csv('../pred/cat_2.csv')
cat3 = pd.read_csv('../pred/cat_3.csv')
cat4 = pd.read_csv('../pred/cat_4.csv')
cat5 = pd.read_csv('../pred/cat_5.csv')

cat = cat1[target_category_str] * 0.2 +  cat2[target_category_str] * 0.2 +  cat3[target_category_str] * 0.2 +  cat4[target_category_str] * 0.2 +  cat5[target_category_str] * 0.2
del cat1, cat2, cat3, cat4, cat5
```

```python
xgb1 = pd.read_csv('../pred/xgb_1.csv')
xgb2 = pd.read_csv('../pred/xgb_2.csv')
xgb3 = pd.read_csv('../pred/xgb_3.csv')
xgb4 = pd.read_csv('../pred/xgb_4.csv')
xgb5 = pd.read_csv('../pred/xgb_5.csv')

xgb = xgb1[target_category_str] * 0.2 +  xgb2[target_category_str] * 0.2 +  xgb3[target_category_str] * 0.2 +  xgb4[target_category_str] * 0.2 +  xgb5[target_category_str] * 0.2
del xgb1, xgb2, xgb3, xgb4, xgb5
```

```python
nn = pd.read_csv('../pred/nn_1024.csv')
```

```python
ens = lgb[target_category_str] * 0.2 + cat[target_category_str] * 0.6+ xgb[target_category_str] * 0.2 + nn[target_category_str] * 0.1
ens.head()
```

```python
import datetime

now = datetime.datetime.now()
now = now.strftime("%m%d_%H%M")
ens.round(5).to_csv(f'../pred/finalsubmission_{now}.csv', index=None)
```
